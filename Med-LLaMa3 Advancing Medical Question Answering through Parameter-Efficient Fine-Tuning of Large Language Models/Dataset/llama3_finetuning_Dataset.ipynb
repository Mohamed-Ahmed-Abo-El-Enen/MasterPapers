```python
from datasets import load_dataset
from datasets import concatenate_datasets
from datasets import *

def mmlu_datasets(split_list):
    def select_answer(sample):
        return {"choices": "\n".join([f"{num+1}) {answer}"for num, answer in enumerate(sample["choices"])]), "type": "MCQ", "context": ""}

    def increment_answer(sample):
        return {"output": str(sample["answer"]+1)}

    def concated_dataset (dataset_id, split_list, name=""):
        dataset = load_dataset(dataset_id, name=name, split=split_list)
        return concatenate_datasets(dataset)

    dataset_1 = concated_dataset("cais/mmlu", split_list, name="clinical_knowledge")
    dataset_2 = concated_dataset("cais/mmlu", split_list, name="college_biology")
    dataset_3 = concated_dataset("cais/mmlu", split_list, name="college_medicine")
    dataset_4 = concated_dataset("cais/mmlu", split_list, name="medical_genetics")
    dataset_5 = concated_dataset("cais/mmlu", split_list, name="professional_medicine")
    dataset_6 = concated_dataset("cais/mmlu", split_list, name="anatomy")

    dataset = [dataset_1, dataset_2, dataset_3, dataset_4, dataset_5, dataset_6]
    dataset = concatenate_datasets(dataset)
    dataset = dataset.select_columns(["question", "choices", "answer"])
    dataset = dataset.rename_column("question", "input")
    dataset = dataset.map(lambda sample: select_answer(sample))
    dataset = dataset.map(lambda sample: increment_answer(sample), remove_columns=["answer"])
    empty_context = ["Choose the correct answer for the following question"] * len(dataset)
    dataset = dataset.add_column("instruction", empty_context)

    return dataset

dataset_4 = mmlu_datasets(split_list=["dev", "validation"]) #, "test"])

print(dataset_4)
dataset_4.to_pandas().sample(5)
```

```python
from datasets import load_dataset
import ast

def medalpaca_medical_meadow_mmmlu_dataset(split_list):
    def concated_dataset (dataset_id, split_list, name=""):
        dataset = load_dataset(dataset_id, name=name, split=split_list)
        return concatenate_datasets(dataset)

    def find_between( s, first, last ):
        try:
            start = s.index( first ) + len( first )
            end = s.index( last, start )
            return s[start:end]
        except ValueError:
            return ""

    def generate_choices(sample):
        sentance = sample["input"].replace('??', '?').replace('? ?', '?').split('?')
        choices = find_between(sentance[1], "{", "}")
        choices = ast.literal_eval('{'+choices+'}')

        keys = list(choices.keys())
        values = list(choices.values())
        choices = ".\n".join([f"{k}) {v}"for k, v in zip(keys, values)])

        question = sentance[0].split(':')[1] + '?'

        answer = sample["output"].split(':')[0]

        return {"input": question, "output": answer, "choices": choices, "type": "MCQ", "context": ""}

    dataset = concated_dataset("medalpaca/medical_meadow_medqa", split_list)
    dataset = dataset.map(lambda sample: generate_choices(sample))
    return dataset

dataset_10 = medalpaca_medical_meadow_mmmlu_dataset(split_list=["train"])

print(dataset_10)
dataset_10.to_pandas().sample(5)
```

```python
from transformers import AutoTokenizer

model_id = "meta-llama/Llama-3.1-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id,
                                          device_map="auto",
                                          trust_remote_code=True
                                          )

tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = 'left'
```

```python
import multiprocessing


def optimal_workers():
    num_cpus = multiprocessing.cpu_count()
    num_gpus = torch.cuda.device_count()
    optimal_value = min(num_cpus, num_gpus*4) if num_gpus else num_cpus - 1
    return optimal_value

num_cpus = optimal_workers()

def prompt_Questions_Type(data_point):
    system = "You are a Medical Assistant follow the following instruction"

    if len(data_point["context"]) >= 5:
        instruction = f"""
{data_point["instruction"]}

{data_point["context"]}
"""

    else:
        instruction = f"""
{data_point["instruction"]}
"""

    if len(data_point["choices"]) > 0:
        prompt_template ="""{instruction}

{input}

{choices}
"""
        prompt = prompt_template.format(instruction=instruction.strip(),
                                        input=data_point["input"].strip(),
                                        choices=data_point["choices"].strip())
    else:
        prompt_template = """{instruction}

{input}
"""

        prompt = prompt_template.format(instruction=data_point["instruction"].strip(),
                                        input=data_point["input"].strip())

    message = [{"role": "system", "content": system},
               {"role": "user", "content": prompt},
               {"role": "assistant", "content": data_point["output"].strip()}]

    return message


def format_instruct_prompt(data_point, tokenizer):
    message = prompt_Questions_Type(data_point)
    data_point["prompt"] = tokenizer.apply_chat_template(message,
                                                        tokenize=False,
                                                        max_length = max_seq_length,
                                                        truncation=True)

    if len(tokenizer(data_point["prompt"]).input_ids) > max_seq_length:
        data_point["prompt"] = None

    return data_point

def generate_prompt(model_id, dataset):
    tokenizer = AutoTokenizer.from_pretrained(model_id,
                                            device_map="auto",
                                            trust_remote_code=True
                                            )

    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = 'left'

    # column_names = list(dataset.features)
    dataset = dataset.map(format_instruct_prompt,
                          num_proc=num_cpus,
                          fn_kwargs={"tokenizer": tokenizer},
                          # remove_columns=column_names,
                          desc="Applying chat template", )
    return dataset
```
